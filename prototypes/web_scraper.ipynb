````python
def web_scrape(filtered_jobs) -> List[Job]:

    scrapped_jobs = []
    for i, job in enumerate(filtered_jobs):
        if not job.text:
            try: 
                response = requests.get(job.url, timeout=5)

                if response.status_code != 200:
                    print(f"JOB ID {i + 1}: Failed with status {response.status_code}")
                else:
                    soup = BeautifulSoup(response.text, "lxml")
                    text = soup.get_text(separator="\n\n", strip=True)

                    if text:
                        # Assign the scraped text into the text attribute of Job object
                        job.text = text
                    
            except requests.exceptions.RequestException as e:
                print(f"JOB ID {i + 1}: Failed to scrape - {e}")

            # Small random delay (always)
            time.sleep(random.uniform(0.8, 1.3))

            # Batch pause (longer delay) every 15 jobs
            if (i + 1) % 15 == 0:
                time.sleep(random.uniform(1.5, 2.0))

        scrapped_jobs.append(job)
                
    return scrapped_jobs
````

## Filter Logic Prototype
To determine which job to pass to the fall-back AI auto browsing agent:
- if False, head to the next step
- if True, send to the fallback auto-browsing logic

````python
def is_low_quality(jobs: List[Job]) -> List[Job]:
    """
    Check if job.text (job description) is at low quality which implies an improper web scrape or poor content.
    If determined as poor content, the Job objects (with low_quality=True) should pass to the auto-browsing class later on.
    """

    # Missing job related keywords
    keywords = ["responsibilities", "requirements", "experience", "role", "job"]
    counter = 0
    for job in jobs:

        text = job.text or ""
        words = text.split()
        low_quality = False
        
        if len(words) < 150:
            low_quality = True
    
        if not any(k in text.lower() for k in keywords):
            low_quality = True

        job.low_quality = low_quality

        if job.low_quality:
            counter += 1

    print(f"LOW QUALITY URLs: {counter} out of {len(jobs)} URLs")
    
    return jobs
````
