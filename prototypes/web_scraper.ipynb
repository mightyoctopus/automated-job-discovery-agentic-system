## Check Root Domain Order Arangement
To check if repeat domains are arranged consicutively (same doamin for prev and current) in order to avoid domain request abuse in web-scraping


````python
from urllib.parse import urlparse
````

````python
def show_domain_list(jobs): 
    domain_list = [job.url for job in all_jobs]
    for domain in domain_list:
        print(domain)
````

````python
### Original Domain Ordering
show_domain_list(all_jobs)
````

````python
def no_adjacent_same_domain(jobs, max_attempts=50):
    def get_root_domain(url):
        return urlparse(url).netloc

    for _ in range(max_attempts):
        random.shuffle(jobs)
        valid = True

        for i in range(len(jobs) - 1):
            if get_root_domain(jobs[i].url) == get_root_domain(jobs[i + 1].url):
                valid = False
                break

        if valid:
            return jobs

    # fallback (important)
    return jobs
````

## Web Scrape

````python
def web_scrape(filtered_jobs) -> List[Job]:

    scrapped_jobs = []
    for i, job in enumerate(filtered_jobs):
        if not job.text:
            try: 
                response = requests.get(job.url, timeout=5)

                if response.status_code != 200:
                    print(f"JOB ID {i + 1}: Failed with status {response.status_code}")
                else:
                    soup = BeautifulSoup(response.text, "lxml")
                    text = soup.get_text(separator="\n\n", strip=True)

                    if text:
                        # Assign the scraped text into the text attribute of Job object
                        job.text = text
                    
            except requests.exceptions.RequestException as e:
                print(f"JOB ID {i + 1}: Failed to scrape - {e}")

            # Small random delay (always)
            time.sleep(random.uniform(1, 1.5))

            # Batch pause (longer delay) every 15 jobs
            if (i + 1) % 15 == 0:
                time.sleep(random.uniform(2, 3))

        scrapped_jobs.append(job)
                
    return scrapped_jobs
````

````python
def check_jd_quality(jobs: List[Job]) -> List[Job]:
    """
    Check if job.text (job description) is at low quality which implies an improper web scrape or poor content.
    If determined as poor content, the Job objects (with low_quality=True) should pass to the auto-browsing class later on.
    """

    # Missing job related keywords
    keywords = [
    "responsibilities", "requirements", "experience",
    "qualifications", "skills", "what you'll do", "about the role"
    ]
    counter = 0
    for job in jobs:

        text = job.text or ""
        words = text.split()
        low_quality = False
        
        if len(words) < 150:
            low_quality = True
    
        if not any(k in text.lower() for k in keywords):
            low_quality = True

        job.low_quality = low_quality

        if job.low_quality:
            counter += 1

    print(f"LOW QUALITY URLs: {counter} out of {len(jobs)} URLs")
    
    return jobs
````
